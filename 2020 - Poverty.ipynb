```python
import pandas as pd
import sqlite3 as sql
from pathlib import Path
import requests
# from IPython.display import Image
# import config

# census_api_key = '&key=' + config.census_api_key

poverty_2020 = 'https://api.census.gov/data/timeseries/poverty/saipe?get=NAME,SAEPOVRTALL_PT&for=COUNTY:*&time=2020'

table_maker = poverty_2020

response = requests.get(table_maker)
print(response)
res = response.json()
res = pd.DataFrame(res)
res['FIPS'] = res[3] + res[4]
res = res.rename(columns = {0:'County Name', 1:'All ages in poverty Rate Estimate', 2:'Year', 3:'Sate', 4:'County', 5:'FIPS'})
res = res.drop(res.index[0])
file_name = 'poverty_2020'
filepath = r'..\cpst_tables\tbl-' + file_name + '.csv'
res.to_csv(filepath, index=False)
```
