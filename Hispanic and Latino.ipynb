```python
# This is for 2010

import pandas as pd
import sqlite3 as sql
from pathlib import Path
import requests

api_summary_file_2010 = 'http://api.census.gov/data/2010/dec/sf1?'

# PCT011001 = Total Population
# PCT011002 = Not Hispanic or Latino
# PCT011003 = Hispanic or Latino

hispanic_latino = api_summary_file_2010 + 'get=PCT011001,PCT011002,PCT011003&for=COUNTY:*'
table_maker = hispanic_latino

response = requests.get(table_maker)
res = response.json()
res = pd.DataFrame(res)
print(response)
res
```

```python
res['FIPS'] = res[3] + res[4]
res = res.rename(columns = {0:'Total population', 1:'Not Hispanic or Latino', 2:'Hispanic or Latino', 3:'State', 4:'County', 5:'FIPS'})
res = res.drop(res.index[0])
file_name = 'hispanic_latino_2010'
filepath = r'..\cpst_tables\tbl-' + file_name + '.csv'
res.to_csv(filepath, index=False)
```
