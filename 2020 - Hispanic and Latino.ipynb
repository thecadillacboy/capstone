```python
import pandas as pd
import sqlite3 as sql
from pathlib import Path
import requests

api_summary_file_2020 = 'https://api.census.gov/data/2020/dec/pl'

# PCT011001 = Total Population
# PCT011002 = Not Hispanic or Latino
# PCT011003 = Hispanic or Latino

hispanic_latino = api_summary_file_2020 + '?get=P2_001N,P2_002N,P2_003N&for=COUNTY:*'
table_maker = hispanic_latino


table_maker = hispanic_latino

response = requests.get(table_maker)
res = response.json()
res = pd.DataFrame(res)
print(response)
res
```

```python
res['FIPS'] = res[3] + res[4]
res = res.rename(columns = {0:'Total population', 1:'Hispanic or Latino', 2:'Not Hispanic or Latino', 3:'State', 4:'County', 5:'FIPS'})
res = res.drop(res.index[0])
file_name = 'hispanic_latino_2020'
filepath = r'..\cpst_tables\tbl-' + file_name + '.csv'
res.to_csv(filepath, index=False)
```
