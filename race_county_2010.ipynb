```python
import pandas as pd
import sqlite3 as sql
from pathlib import Path
import requests
# from IPython.display import Image
# import config

# census_api_key = '&key=' + config.census_api_key

api_summary_file_2010 = 'http://api.census.gov/data/2010/dec/sf1?'
county_pop_2010 = api_summary_file_2010 + 'get=PCT023001,PCT023003,PCT023004,PCT023005,PCT023006,PCT023014,PCT023019,PCT023020&for=COUNTY:*'

table_maker = county_pop_2010

response = requests.get(table_maker)
print(response)
```

```python
# PCT023001 = Total population
# PCT023003 = White
# PCT023004 = Black or African American
# PCT023005 = American Indian and Alaska Native
# PCT023006 = Asian
# PCT023014 = Native Hawaiian and Other Pacific Islander
# PCT023019 = Some Other Race
# PCT023020 = Two or More Races
```

```python
res = response.json()
res = pd.DataFrame(res)
res['FIPS'] = res[8] + res[9]
res = res.rename(columns = {0:'Total population', 1:'White', 2:'Black or African American', 3:'American Indian and Alaska Native', 4:'Asian', 5:'Native Hawaiian and Other Pacific Islander', 6:'Some Other Race', 7:'Two or More Races', 8:'State', 9:'County', 10:'FIPS'})
res = res.drop(res.index[0])
file_name = 'discard-race_county_2010'
filepath = r'..\cpst_tables\tbl-' + file_name + '.csv'
res.to_csv(filepath, index=False)
res
```
